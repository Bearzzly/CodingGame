```python
# Full Validate by recursion
import sys
import math
import re

# Auto-generated code below aims at helping you parse
# the standard input according to the problem statement.

n = int(input()) #nbr of calc
cel = {}

def compute(cel, dico):
    if cel["sol"] == False:
        for i in ["arg1", "arg2"]:
            if "$" in cel["op"][i]:
                i_value = cel["op"][i].replace("$", "")
                compute(dico[int(i_value)], dico)
                cel["op"][i] = dico[int(i_value)]["res"]

        if cel["op"]["op"] == "VALUE":
            cel["res"] = cel["op"]["arg1"]
        elif cel["op"]["op"] == "ADD":
            cel["res"] = int(cel["op"]["arg1"]) + int(cel["op"]["arg2"])
        elif cel["op"]["op"] == "SUB":
            cel["res"] = int(cel["op"]["arg1"]) - int(cel["op"]["arg2"])
        elif cel["op"]["op"] == "MULT":
            cel["res"] = int(cel["op"]["arg1"]) * int(cel["op"]["arg2"])

        cel["sol"] = True

for i in range(n):
    op, arg1, arg2 = input().split()
    cel[i] = { "res" : None,
                "op" : {
                    "op" : op,
                    "arg1" : arg1,
                    "arg2" : arg2
                },
                "sol": False
    }

for i in range(n):
    compute(cel[i], cel)
    print(cel[i]["res"])

#1
#print(3)
#print(7)

#2
#print("20")
#print("120")
#print("121")

#3
#print(12)
#print(9)

#5
#print(3)
#print(2)
#print(8)

#7 fibonacci
#print(0)
#print(1)
#print(1)
#print(2)
#print(3)
#print(5)
#print(8)
#print(13)
#print(21)
```
