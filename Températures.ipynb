```python
import sys
import math
 
res = int() #declare for first init of  res 
n = int(input())  # the number of temperatures to analyse
isSet = 0  #check if first init of res already done
for i in input().split():
    t = int(i)
    if isSet == 0:
        res = t
        isSet = 1
    if abs(t) == abs(res): #check absolue value, no more negative
        if t >= 0:
            res = t    #keep the closest from 0 but positiv
    elif abs(t) < abs(res):
        res = t

if n == 0 :
    print(0)
else :
    print(res)
```
